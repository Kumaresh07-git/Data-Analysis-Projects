```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("ecommerce_sales.csv")
df.head(20)
```

```python
df.dtypes
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df.dtypes
```

```python
df['order_month'] = df['order_date'].dt.month_name()
df['order_year'] = df['order_date'].dt.year
df.head(10)
```

```python
df.info()
df.describe(include = 'all')
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data = df, x = 'category', order = df['category'].value_counts().index)
plt.title('Number of Orders by Category')
plt.show()

sns.countplot(data = df, x = 'region', order = df['region'].value_counts().index)
plt.title('Number of Orders by Region')
plt.show()
```

```python
new_cols = ['discount', 'total_amount', 'price', 'quantity']

for cols in new_cols:
    plt.figure(figsize=(8,5))
    sns.histplot(df[cols], bins = 30, kde = True)
    plt.title(f"Distribution of {cols}")
    plt.show()
```

```python
plt.figure(figsize = (8, 5))
sns.scatterplot(data = df, x = 'discount', y = 'profit_margin', alpha = 0.6)
plt.title('Discount vs Profit Margin')
plt.show()
```

```python
category_revenue = df.groupby('category')['total_amount'].sum().sort_values(ascending = False)
plt.figure(figsize = (8, 5))
sns.barplot(x = category_revenue.index, y = category_revenue.values)
plt.title('Total Revenue by Category')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data = df, x = 'returned', y = 'profit_margin')
plt.title('Profit Margin by Returned Status')
plt.show()
```

```python
plt.figure(figsize = (8, 5))
payment_counts = df['payment_method'].value_counts()
plt.pie(payment_counts,
       labels = payment_counts.index,
       autopct = '%1.1f%%',
       startangle = 90,
       explode = [0.05] * len(payment_counts)
)
plt.title('Payment Menthod')
plt.show()
```

```python
plt.figure(figsize = (8, 5))
gender_counts = df['customer_gender'].value_counts()
plt.pie(gender_counts,
       labels = gender_counts.index,
       autopct = '%1.1f%%',
       startangle = 90,
)
plt.show()
```

```python
bins = [17, 25, 40, 100]
labels = ['18-25', '26-40', '41-60+']

df['age_group'] = pd.cut(df['customer_age'], bins = bins, labels = labels)
```

```python
age_payment = pd.crosstab(df['age_group'], df['payment_method'])
age_payment
```

```python
ax = age_payment.plot(
    kind = 'bar',
    figsize = (10, 5),
    colormap = 'viridis',
    width = 0.8
)
plt.title('Payment Method Distribution by age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Orders')
plt.legend(title = 'Payment Method')


for container in ax.containers:
    ax.bar_label(container, label_type = 'center', fontsize = 9, color = 'white', rotation = 90)
plt.show()
```
